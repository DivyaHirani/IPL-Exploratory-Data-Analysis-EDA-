4.Exploratory Data Analysis (EDA)

  Perform exploratory analysis using Pandas, Matplotlib, and Seaborn.



Summary statistics:

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

country_counts = df['Country'].value_counts()

country_counts.plot(kind='bar', color=['blue', 'red'])
plt.title("Distribution of Players by Country")
plt.xlabel("Country")
plt.ylabel("Number of Players")
plt.show()
```

Top 5 genres with most releases:

```python
from collections import Counter
genres = df['Country'].dropna().str.split(', ')
genre_counts = Counter([genre for sublist in genres for genre in sublist])
top_genres = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)
top_genres.head()
```

country data as per year:

```python
df['Country'].value_counts().plot(kind='line', figsize=(10, 5))
plt.title("Player Count by Country")
plt.xlabel("Country")
plt.ylabel("Number of Players") .show()
```

```python
top_players = df['Player_Name'].value_counts().head(10)
top_players.plot(kind='pie', color='green')
plt.title("Top 10 players ")
plt.show()
```
